# STATS

```python
class_paths = [f.path for f in os.scandir(data_root) if f.is_dir()]
```

```python
for c in class_paths:
    subcategory_paths = [f.path for f in os.scandir(c) if f.is_dir()]
    subcategories = [os.path.basename(os.path.normpath(x)) for x in subcategory_paths]
    print(subcategories)
```

```python
d = []
for c in class_paths:
    cat = os.path.basename(os.path.normpath(c))
    subcategory_paths = [f.path for f in os.scandir(c) if f.is_dir()]
    for subc in subcategory_paths:
        subcat = os.path.basename(os.path.normpath(subc))
        division_paths = [f.path for f in os.scandir(subc) if f.is_dir()]
        for div in division_paths:
            division = os.path.basename(os.path.normpath(div))
            _, _, files = next(os.walk(div))
            d.append({"class": cat, "subcategory":subcat, "window":division, "samples":len(files)})
```

```python
df = pd.DataFrame(d)
```

# Creating new clean folder

```python
clean_root = '.\\..\\datasets\\data_clean_xlarge'
```

```python
counter = dict()
for c in class_paths:
    cat = os.path.basename(os.path.normpath(c))
    counter[cat] = 0
    os.makedirs(os.path.join(clean_root, cat))
    subcategory_paths = [f.path for f in os.scandir(c) if f.is_dir()]
    for subc in subcategory_paths:
        subcat = os.path.basename(os.path.normpath(subc))
        division_paths = [f.path for f in os.scandir(subc) if f.is_dir()]
        for div in division_paths:
            division = os.path.basename(os.path.normpath(div))
            _, _, files = next(os.walk(div))
            for file in files:
                im = Image.open(os.path.join(div, file))
                filemame = cat + '_' + subcat + '_' + division + '_' + str(counter[cat]) + '.jpg'
                if 'png' in file:
                    im = im.convert('RGB')
                im.save(os.path.join(clean_root, cat, filemame))
                counter[cat] += 1
```
